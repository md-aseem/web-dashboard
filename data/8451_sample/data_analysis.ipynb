```python
import pandas as pd
import os

hhd_path = r"C:\Work\Courses\Intro to Cloud Computing\myflaskapp\data\8451_sample\400_households.csv"
pd_path = r"C:\Work\Courses\Intro to Cloud Computing\myflaskapp\data\8451_sample\400_products.csv"
tr_path = r"C:\Work\Courses\Intro to Cloud Computing\myflaskapp\data\8451_sample\400_transactions.csv"

# Load data
hhd_df = pd.read_csv(hhd_path).rename(columns=lambda x: x.strip())
pd_df = pd.read_csv(pd_path).rename(columns=lambda x: x.strip())
tr_df = pd.read_csv(tr_path).rename(columns=lambda x: x.strip())

combined_df = pd.merge(tr_df, pd_df, on='PRODUCT_NUM', how='left')
combined_df = pd.merge(combined_df, hhd_df, on='HSHD_NUM', how='left')
combined_df['Month'] = pd.to_datetime(combined_df['PURCHASE_'], format='%d-%b-%y').dt.month_name()

# Filter for Household Number 10 (as a sample) and sort data
filtered_sorted_df = combined_df.sort_values(by=['HSHD_NUM', 'BASKET_NUM', 'PURCHASE_', 'PRODUCT_NUM', 'DEPARTMENT', 'COMMODITY'])

# Display the result
(filtered_sorted_df['HSHD_NUM'].unique().sort())
filtered_sorted_df
```
